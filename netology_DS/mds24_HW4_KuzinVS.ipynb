```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

#Домашнее задание "Функции одной переменной".
##Уровень 0:
###Задание 1
Посчитайте значение первой и второй производной функции.
В точке x = 1.

```python
def funct(x):
  return x**5 + 4 * sym.sin(2 * x) + sym.cos(3 * x + 3)

x = sym.Symbol('x')
y1 = sym.diff(funct(x)).evalf(subs={'x':1})
y2 = sym.diff(sym.diff(funct(x))).evalf(subs={'x':1})
print(y1)
print(y2)
```

###Задание 2
Постройте график первой и второй производной функции:

f(x) = sin**5 * (2 * x + 1)

На отрезке [-5; 5].

```python
fx0 = sym.sin(2 * x + 1)**5

def f1(x):
    return sym.diff(fx0).evalf(subs={'x': x})

def f2(x):
    return sym.diff(sym.diff(fx0)).evalf(subs={'x': x})

x = sym.Symbol('x')

x1 = np.linspace(-5, 5, 100)
y1 = [f1(x1[i]) for i in range(len(x1))]
y2 = [f2(x1[i]) for i in range(len(x1))]
```

```python
plt.plot(x1, y1)
plt.plot(x1, y2)
plt.xlabel('x')
plt.ylabel('y')
```
